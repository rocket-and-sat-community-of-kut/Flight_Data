```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
press_temp_log = np.genfromtxt('Press_Temp.csv', delimiter=',')
Time_log = press_temp_log[:, 6] # [s]
Press_log = press_temp_log[:,7] # [hPa]
Temp_log = press_temp_log[:,8] #[℃]
Alt_log = press_temp_log[:,10] #[m]

Roll_log = np.genfromtxt('Roll.csv', delimiter=',')
Roll_time_log = np.arange(0,1/50*781,1/50)
Encoder_log = Roll_log[:,1]
Gyro_log = Roll_log[:,2]
Target_log = Roll_log[:,3]
Duty_log = Roll_log[:,4]

fig,ax = plt.subplots(2,2,figsize=(16, 12))
ax[0,0].set_xlabel("Time[s]")
ax[0,0].set_ylabel("Pressure[hPa]")
ax[0,0].grid()
ax[0,0].plot(Time_log,Press_log,color="C0")

ax[0,1].set_xlabel("Time[s]")
ax[0,1].set_ylabel("Temperature[℃]")
ax[0,1].grid()
ax[0,1].plot(Time_log,Temp_log,color="C1")

ax[1,0].set_xlabel("Time[s]")
ax[1,0].set_ylabel("Altitude[m]")
ax[1,0].grid()
ax[1,0].plot(Time_log,Alt_log,color="C2")

ax[1,1].set_xlabel("Time[s]")
ax[1,1].set_ylabel("Angular Velocity[deg/s]")
ax[1,1].grid()
ax[1,1].plot(Roll_time_log,Encoder_log,color="C3",label="Encoder")
ax[1,1].plot(Roll_time_log,Gyro_log,color="C4",label="Gyro")
plt.show()
```
